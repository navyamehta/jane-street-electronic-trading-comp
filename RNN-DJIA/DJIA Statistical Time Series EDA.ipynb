```python
import pandas as pd
import numpy as np
import scipy as sp
import os
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.diagnostic import acorr_ljungbox
```

<h2> Assimilate Stock Data </h2>

```python
data = pd.DataFrame()
for source, dirs, files in os.walk("./data/rawdata"):
    for file in files:
        df = pd.read_csv(os.path.join(source, file))
        if len(df)==0: continue
        df = df.set_index("Date")["Close"].rename(df["Name"].values[0])
        data = df if len(data)==0 else pd.concat([data, df], axis=1)
data = data.ffill()
```

```python
initcapital = 7500
weights = data.apply(lambda s: s/np.sum(s), axis=1)
djia = (data/data.shift(1)-1)*weights.shift(1)
djia = (djia.apply(np.sum, axis=1)+1).cumprod()*initcapital
```

```python
plt.figure(figsize=(15,4))
djia.plot()
plt.grid(which="both", axis="both", visible=True, linestyle="--")
plt.ylabel("DJIA Capital Level")
plt.title("Simulated DJIA Price-Weighted Values")
plt.show()
```

<h2> Dissect Time Series Characteristics </h2>

```python
#Force Stationarity - Augmented Dickey Fuller Test
adf = pd.Series(np.nan, np.arange(3, dtype=np.int))
temp, best = djia.copy(), djia.copy()
setter = True
for i in adf.index:
    adf[i] = adfuller(temp.dropna())[1]
    if (adf[i]<0.05) and setter:
        best = temp.copy()
        setter=False
    temp = temp/temp.shift(1)
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
best.plot()
ax = plt.subplot(1,2,2)
ax.axis('tight')
ax.axis('off')
plt.title("P-Values for Augmented Dickey Fuller Test")
ax.table(cellText=adf.to_frame().reset_index().values, colLabels=["Differencing", "P-Value"], loc="center")
```

```python
#Serial Correlation - Ljung-Box Test
def autocor(series):
    pcor = pd.Series(np.arange(series.shape[0]/3, dtype=np.int), np.arange(series.shape[0]/3))
    pcor = pcor.apply(lambda s: np.corrcoef(series.iloc[s:], series.shift(s).dropna())[0][1])
    return pcor
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.title("Average Ljung-Box P-Value = "+str(acorr_ljungbox(djia)[1].mean()))
autocor(djia).plot()
plt.xlabel("Lag")
plt.ylabel("Pearson's Correlation Coefficient")
plt.subplot(1,2,2)
plt.title("Average Ljung-Box P-Value = "+str(acorr_ljungbox(best.dropna())[1].mean()))
autocor(best.dropna()).plot()
plt.xlabel("Lag")
plt.ylabel("Pearson's Correlation Coefficient")
plt.show()
```

```python
#Fourier Transform Analysis
cmplx = sp.fftpack.fft(best.dropna())[1:]
#Amplitude-Frequency
plt.bar(np.arange(len(cmplx)), np.abs(cmplx))
plt.title("Amplitude vs Frequency for Consistuent Sinusoids")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
